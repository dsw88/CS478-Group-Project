```python
import pandas
import numpy as np
df=pandas.read_csv("combined.csv")
features = df[['make', 'model', 'mileage', 'state', 'transmission', 'year','price']]
```

```python
f=open("combined_data.arff","w")
df=df.replace(np.nan,"?")
df=df.drop( ['vin'],axis=1)
df=df.replace('Not Specified',"?")
df=df.replace('not specified',"?")
df=df.replace('nan',"?")
columns=df.columns
f.write("@RELATION cars\n")
for column in columns:
    f.write("@ATTRIBUTE ")
    f.write("'"+str(column)+"'")
    unique=df[column].unique()
    unique=map(str, unique)
    unique=set(unique)
    if(len(unique)>100) and column!='trim' and column!='color'and column!='model' :
        f.write(" NUMERIC\n")
    else:
        f.write(" {")
        f.write(', '.join([("'"+str(x)+"'" if not x.isdigit() else x) for x in unique]))
        f.write("}\n")
f.write("@DATA\n")
print(df)
for index,row in df.iterrows():
    row=map(str, row)
    f.write(', '.join([("'"+str(x)+"'" if not ((str(x)).isdigit() or x=='?') else x) for x in row]))
    f.write('\n')
            
f.close()
```
